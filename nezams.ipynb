```python
import requests
```

```python
def convert_string(string):
    stripped_string = string.strip().replace(' ', '-')
    return stripped_string


def first_four(string):
    stripped_string = string.strip().replace(' ', '-')
    four_words = string.split()[:4]
    join_string = '-'.join(four_words)
    return join_string
```

```python
base_url = "https://nezams.com/"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.93 Safari/537.36"
}
```

```python
#note that 2 and 5 arent working

for titel in to_scrape:
    url = base_url + convert_string(titel)
    response = requests.get(url, headers=headers)
    print(response)
    print(url)
```

```python
# it works with all the required url
laws = []
urls = []
for titel in to_scrape:
    url = base_url + first_four(titel)
    response = requests.get(url, headers=headers)
    print(response)
    urls.append(url)
    print(url)
    soup = bs4(response.text)
    print(soup.find('h1').text)
    laws.append(soup)
```

Law Table (csv):
- law_id (int): unique id for the law
- law_title (str): title of the law
- law_url (str): url of the law
- law_status (str): status of the law (النفاذ)
- law_category (str): category of the law (التصنيف)

Article Table (csv):
- law_id (int): law_id of the law that the article belongs to
- article_number (int): number of the article from (المادة الأولى, المادة الثانية,...)
- article_text (str): text of the article
